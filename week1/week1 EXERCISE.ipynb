````python
# constants

MODEL_GPT = 'gpt-4o-mini'
#MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
system_prompt="you act as an technical assistant who provides in brief clear and exact solution the question asked by user"
system_prompt+="the answer should be crisp and easy to understand."

def user_prompt(query):
    user=f"Provide me an solution to this query {query}"
    user+="answer should be simple and display output in markdown"
    return user
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

````python
# Get gpt-4o-mini to answer, with streaming
import openai
def tech_tool(query):
    result=openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {'role':"system","content":system_prompt},
            {'role':"user","content":user_prompt(query)}
        ],
        stream=True)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in result:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
tech_tool(question)
````
